```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
f = open('C:/Users/79632/source/repos/Laba_4_MATMOD/result.txt', 'r')

a, b = map(float, f.readline().split())

txyz = []
for point in f:
    txyz.append(list(map(float, f.readline().split())))

txyz = np.array(txyz)
print(txyz.shape)
print(txyz)
fig, axs = plt.subplots(2, 2, figsize=(16,10))
axs[0,0].plot(txyz[:, 1], txyz[:, 2])
axs[0,0].set_xlabel('x')
axs[0,0].set_ylabel('y')
axs[0,1].plot(txyz[:, 2], txyz[:, 3])
axs[0,1].set_xlabel('y')
axs[0,1].set_ylabel('z')
axs[1,0].plot(txyz[:, 1], txyz[:, 3])
axs[1,0].set_xlabel('x')
axs[1,0].set_ylabel('z')
axs[1,1].plot(txyz[:, 0], txyz[:, 1])
axs[1,1].set_xlabel('t')
axs[1,1].set_ylabel('x')
plt.show()



fig = plt.figure(figsize=(16, 10))
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# plotting
ax.plot3D(txyz[:, 1], txyz[:, 2], txyz[:, 3], 'green')
plt.show()
```
